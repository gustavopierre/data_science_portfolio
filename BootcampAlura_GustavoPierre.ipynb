```
#outra forma de ler a tabela, copiando como uma variável string
from io import  StringIO

dados_da_populacao_io = StringIO(dados_da_populacao)

pop = pd.read_csv(dados_da_populacao_io, sep='\t')
```

```
def limpeza(populacao, gastos):
    # apagando linhas com dados nulos
    populacao = populacao.dropna()
    #arracando os espaços da coluna populacao e transformando-a em inteiro
    populacao['populacao'] = populacao['População'].str.replace(" ", "").astype(int)
    populacao['uf'] = populacao['Unidade federativa']

    # extraindo só as colunas uf e populacao e definindo o índice como sendo uf
    populacao = populacao[['uf','populacao']].set_index('uf')
    # limpando os espaços vazios antes e depois dos nomes dos estado
    populacao.index = populacao.index.str.strip()
    # excluidndo o código do estado do nome do estado
    #gastos.index = gastos.index.str[3:]
    #excluindo os nomes dobrados de estado
    for estado in gastos.index:
        populacao.index = populacao.index.str.replace(f'{estado} {estado}', estado)
    
    unida = populacao.join(gastos)
    return populacao, unida
```
